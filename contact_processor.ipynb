````python
import pandas as pd
import secrets
import string

# Load your data
df = pd.read_csv('raw_contacts.csv') 

def generate_hashtag_id(length=9):
    # Generates a string like #A1B2C3D4E
    chars = string.ascii_uppercase + string.digits
    random_str = ''.join(secrets.choice(chars) for _ in range(length))
    return f"```{random_str}#```"

# Add the column using a list comprehension for speed
df['id'] = [generate_hashtag_id() for _ in range(len(df))]
df['id_2'] = [generate_hashtag_id() for _ in range(len(df))]

df.to_csv('contacts_processed.csv', index=False)
````
